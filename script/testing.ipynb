```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import rcParams
from matplotlib import pyplot
# %matplotlib notebook
```

```python
data= pd.read_excel("density.xlsx")
data
```

```python
plt.figure(figsize=(10,10))
plt.style.use('fast')
pyplot.scatter(x=data.Scanspeed, y=data.Power, s=100, edgecolor ='b', alpha=0.8)
plt.xlabel('Scan Speed(mm/s)', size = 15)
plt.ylabel('Power(W)', size = 15)
for i in range(data.shape[0]):
    plt.text(data.Scanspeed[i],data.Power[i],(data.SampleNum[i]), position=(data.Scanspeed[i], data.Power[i]+5), fontsize='medium', ha='left')
    
    
# plt.text(data.Scanspeed[i],data.Power[i],(data.Scanspeed[i], data.Power[i]))
# print(data.shape[1,4])
# print(data.shape[2:])
# print(data.shape[3])
# print(data.shape[4])
# for i in range(data.shape[0]):
#  plt.text(x=data.Scanspeed[i],y=data.Power[i],s=175,alpha=0.5)
print(data.shape[0])
```

```python
plt.figure(figsize=(11,11))
plt.style.use('fivethirtyeight')
plt.xlabel('Scan Speed(mm/s)', size = 15)
plt.ylabel('Power(W)', size = 15)
plt.scatter(data.Scanspeed,data.Power, c = data.Porosity, cmap ='Spectral' , s=250)
for i in range(data.shape[0]):
    plt.text(data.Scanspeed[i],data.Power[i],(round(data.Porosity[i],1)), position=(data.Scanspeed[i], data.Power[i]+5), fontsize='medium',va='bottom', ha='center')
cbar=plt.colorbar(orientation = 'vertical');
cbar.set_label("Porosity%",size = 15)
```
