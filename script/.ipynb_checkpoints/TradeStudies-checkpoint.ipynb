```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import rcParams
from matplotlib import pyplot
# %matplotlib notebook
```

#### FIRST YOU NEED TO READ THE DATA FROM EXCEL OR CSV FILE

```python
data= pd.read_excel("sat_data.xlsx")
data
# print(data.unit)
```

```python
plt.figure(figsize=(10,10))
plt.style.use('default')
pyplot.scatter(x=data.mass, y=data.alt, s=250, edgecolor ='b', alpha=0.8)
plt.xlabel('Mass(kg)', size = 15)
plt.ylabel('Altitude(km)', size = 15)
for i in range(data.shape[0]):
    plt.text(data.mass[i],data.alt[i],str(round(data.unit[i]))+'U', position=(data.mass[i]+0.3, data.alt[i]+1), fontsize='medium', ha='left')
plt.savefig("graph01.png")
# plt.text(data.mass[i],data.alt[i],str(data.size[i])+'U', position=(data.mass[i], data.alt[i]), fontsize='medium', ha='left')
plt.savefig('fgraph01.pdf')
```

```python
plt.figure(figsize=(10,10))
plt.style.use('default')
plt.xlabel('Mass(kg)', size = 25)
plt.ylabel('Alt(km)', size = 25)
plt.scatter(data.mass,data.alt, c = data.power, cmap ='rainbow' , s=450)
for i in range(data.shape[0]):
    plt.text(data.mass[i],data.alt[i],str(round(data.unit[i]))+'U', position=(data.mass[i]+.58, data.alt[i]+5), fontsize='medium',va='bottom', ha='center')
cbar=plt.colorbar(orientation = 'vertical');
cbar.set_label("Power(W)",size = 25)
# plt.savefig("graph02.png")
plt.savefig('power.pdf')
plt.savefig('power.svg')
```

```python
plt.figure(figsize=(10,10))
plt.style.use('default')
plt.xlabel('Mass(kg)', size = 25)
plt.ylabel('Alt(km)', size = 25)
plt.scatter(data.mass,data.alt, c = data.alpha, cmap ='rainbow' , s=450)
for i in range(data.shape[0]):
    plt.text(data.mass[i],data.alt[i],str(round(data.unit[i]))+'U', position=(data.mass[i]+.55, data.alt[i]+6), fontsize='medium',va='bottom', ha='center')
cbar=plt.colorbar(orientation = 'vertical');
cbar.set_label("Inclination(Î±)",size = 25)
plt.savefig("alpha.svg")
plt.savefig('alpha.pdf')
```

```python

fig = plt.figure(figsize=(10, 10))
# Adds subplot on position 1
ax = fig.add_subplot()
# fig, ax = plt.subplots()
x=data.mass
y=data.alt
scatter = ax.scatter(x, y, c=data.unit, cmap ='Accent', edgecolors='white', s=400*data.unit)
legend1 = ax.legend(*scatter.legend_elements(),
                    loc="upper right", title="Unit")
ax.add_artist(legend1)

# for color in ['tab:blue', 'tab:orange', 'tab:green']:
#     n = data.shape[0]
#     x=data.mass
#     y=data.alt
#     scale = 500*data.unit
#     ax.scatter(x, y, c=data.unit, cmap ='Accent',s=scale, label=color,
#                alpha=0.3, edgecolors='none')

# ax.legend()
# ax.grid(True)
# legend1 = ax.legend(*scatter.legend_elements(),
#                     loc="upper right", title="Unit Size")
plt.xlabel('Mass(kg)', size = 15)
plt.ylabel('Alt(km)', size = 15)
ax.add_artist(legend1)
plt.savefig("unit.svg")
plt.savefig('unit.pdf')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.style.use('default')
pyplot.scatter(x=data.mass, y=data.alt, s=1800*data.unit, c=data.unit, cmap ='Accent', alpha=0.6, edgecolors='white', linewidth=2)
plt.xlabel('Mass(kg)', size = 15)
plt.ylabel('Altitude(km)', size = 15)
for i in range(data.shape[0]):
    plt.text(data.mass[i],data.alt[i],str(round(data.unit[i]))+'U', position=(data.mass[i]+0.3, data.alt[i]+1), fontsize='medium', ha='left')
plt.savefig("graph05.svg")
# plt.text(data.mass[i],data.alt[i],str(data.size[i])+'U', position=(data.mass[i], data.alt[i]), fontsize='medium', ha='left')
plt.savefig('fgraph05.pdf')
```

```python
plt.figure(figsize=(10,10))
import seaborn as sns
df=pd.read_excel("sat_data.xlsx")
df.sort_values(by=['down'], inplace=True)
# Some boilerplate to initialise things
# This is where the actual plot gets made
ax = sns.barplot(data=df, x="down", y="power", hue=data.unit,saturation=1)
# Customise some display properties
# ax.set_title('UK election results')
ax.grid(False)
# sns.set_style("darkgrid")
# ax.set_facecolor('gray')
ax.set_ylabel('Power(W)')
ax.set_xlabel('Downlink(MHz)')
ax.set_xticklabels(df["down"].unique().astype(str), rotation='vertical')
# Ask Matplotlib to show it
plt.gcf().subplots_adjust(bottom=0.15)
plt.savefig("down.svg")
# plt.savefig('fgraph06.pdf')fervf 
# plt.show()
sns.set()
plt.figure()
```
